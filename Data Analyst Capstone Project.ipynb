## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
import pandas as pd
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection

# Query to extract data
query = "SELECT Age,WorkWeekHrs FROM master;"

# Fetch data into a DataFrame
df = pd.read_sql(query, conn)

import matplotlib.pyplot as plt

# Plot scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], color='skyblue', alpha=0.5)
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
import matplotlib.pyplot as plt
import pandas as pd
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection

# Query to extract data
query = "SELECT Age,WorkWeekHrs,CodeRevHrs  FROM master;"

# Fetch data into a DataFrame
df = pd.read_sql(query, conn)

# Define bubble sizes based on Age column
bubble_sizes = df['Age'] * 5  # You can adjust the multiplier for better visualization

# Plot bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=bubble_sizes, alpha=0.5)
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
import matplotlib.pyplot as plt
import pandas as pd
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection

# Query to extract data
query = "SELECT Age,WorkWeekHrs,CodeRevHrs  FROM master;"

# Fetch data into a DataFrame
df = pd.read_sql(query, conn)

# Assuming 'top_databases' is a dictionary containing the top 5 databases and their percentages
top_databases = {'MySQL': 25, 'PostgreSQL': 20, 'MongoDB': 15, 'SQLite': 10, 'SQL Server': 5}

# Extract database names and their percentages
database_names = list(top_databases.keys())
percentages = list(top_databases.values())

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=database_names, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

import matplotlib.pyplot as plt

# Filter the DataFrame for the age group 30 to 35
age_group_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate median values for WorkWeekHrs and CodeRevHrs
median_work_week_hrs = age_group_df['WorkWeekHrs'].median()
median_code_rev_hrs = age_group_df['CodeRevHrs'].median()

# Plot stacked chart
plt.figure(figsize=(10, 6))
plt.bar(['30-35'], median_work_week_hrs, label='WorkWeekHrs', color='skyblue')
plt.bar(['30-35'], median_code_rev_hrs, bottom=median_work_week_hrs, label='CodeRevHrs', color='orange')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Age Group')
plt.ylabel('Median Hours')
plt.legend()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
import matplotlib.pyplot as plt
import pandas as pd
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection

# Query to extract data
query = "SELECT Age,WorkWeekHrs,CodeRevHrs, ConvertedComp  FROM master;"

# Fetch data into a DataFrame
df = pd.read_sql(query, conn)

import matplotlib.pyplot as plt

# Filter the DataFrame for the age range 45 to 60
age_range_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calculate median ConvertedComp for each age
median_converted_comp = age_range_df.groupby('Age')['ConvertedComp'].median()

# Plot median ConvertedComp for all ages from 45 to 60
plt.figure(figsize=(10, 6))
plt.plot(median_converted_comp.index, median_converted_comp.values, marker='o', color='blue', linestyle='-')
plt.title('Median ConvertedComp for Ages 45-60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
import matplotlib.pyplot as plt
import pandas as pd
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection

# Query to extract data
query = "SELECT Age,WorkWeekHrs,CodeRevHrs, ConvertedComp, MainBranch  FROM master;"

# Fetch data into a DataFrame
df = pd.read_sql(query, conn)

import matplotlib.pyplot as plt

# Count occurrences of each category in MainBranch column
main_branch_counts = df['MainBranch'].value_counts()

# Plot horizontal bar chart
plt.figure(figsize=(10, 6))
main_branch_counts.plot(kind='barh', color='skyblue')
plt.title('Distribution of MainBranch')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.grid(axis='x')
plt.show()
```
